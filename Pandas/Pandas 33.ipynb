### Nilai Kumulatif pada Pandas Data Frame

### Import Module

```python
import pandas as pd

print(pd.__version__)
```

```python
d={'pemain':['Budi','Joni','Iwan','Budi','Budi','Iwan','Asep','Joni'],
  'goal':[2,1,3,1,1,2,2,3]}

df=pd.DataFrame(d)
df
```

## Nilai Kumulatif 

```python
df['goal'].cumsum().to_frame()
```

```python
df['jumlah_goal_kumulatif']=df['goal'].cumsum()
df
```

```python
df['Jumlah_goal)kumulatif_tiap_pemain']=df.groupby('pemain')['goal'].cumsum()
df
```

```python
df['cummax']=df['goal'].cummax()
df
```

```python
df['cummin']=df['goal'].cummin()
df
```

```python
df['cumprod']=df['goal'].cumprod()
df
```
