### Seleksi baris pada DF dengan Fungsi between()

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows=10
n_cols=5
cols=tuple('ABCDE')

df=pd.DataFrame(np.random.randint(1,10,size=(n_rows,n_cols)),
               columns=cols)

df
```

## Seleksi dengan between()

```python
df[df['A'].between(2,5)]
```

```python
df[df['A'].between(2,5,inclusive=False)]
```
