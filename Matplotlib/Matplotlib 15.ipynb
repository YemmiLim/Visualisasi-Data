## 3D Area Plot/Chart

```python
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
n=np.linspace(-3.0,3.0,100)

x,y =np.meshgrid(n,n)
z=np.sqrt(x**2 +y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## 3D Wire Frame

```python
fig=plt.figure()
ax=plt.axes(projection='3d')

ax.plot_wireframe(x,y,z,color='green')

ax.set_xlabel('x')
ax.set_ylabel('x')
ax.set_zlabel('x')
ax.set_title('3D Wireframe')

plt.show()
```

## 3D Surface Plot

```python
fig=plt.figure()
ax=plt.axes(projection='3d')

#ax.plot_surface(x,y,z,color='green')
ax.plot_surface(x,y,z,cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('x')
ax.set_zlabel('x')
ax.set_title('3D Surfaceplot')

plt.show()
```
